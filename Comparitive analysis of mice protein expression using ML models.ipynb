```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Step 1: Download and Load the Dataset
# Assuming the dataset is saved as 'MiceProteinExpression.csv'
data = pd.read_csv('/content/Data_Cortex_Nuclear.csv')

# Drop rows with missing values
data.dropna(inplace=True)

# Fill missing values with mean of other columns
data.fillna(data.mean(), inplace=True)
print(data.to_string())
```

```python
# Step 2: Data Normalization
# Select the appropriate columns for training and testing (excluding non-feature columns)
X = data.drop(['MouseID', 'Genotype', 'Treatment', 'Behavior', 'class'], axis=1)
y = data['class']
print(X.to_string())
print(y.to_string())
```

```python
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
# Step 3: Convert Class Labels to Numerical Values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
print(y_encoded)
# Step 4: Splitting the Dataset into Training and Validation Sets
X_train, X_val, y_train, y_val = train_test_split(X_normalized, y_encoded, test_size=0.2, random_state=42)
```

```python
print("the training data is ",X_train)
print("the traing output is ",y_train)
print("the testing data is ",X_val)
print("the testing data output is ",y_val)
```

```python
# Step 5: Creating the Random Forest Model
rf_model = RandomForestClassifier(n_estimators=90, random_state=42)

# Training the model
rf_model.fit(X_train, y_train)
```

```python
#Step 6: Testing Accuracy on Validation Set
y_pred_val = rf_model.predict(X_val)
y_pred_val_labels = label_encoder.inverse_transform(y_pred_val)
y_val_labels = label_encoder.inverse_transform(y_val)
print(X_val)
print(y_pred_val)
```

```python
val_accuracy = accuracy_score(y_val_labels, y_pred_val_labels)
print("Validation Accuracy:", val_accuracy)
# Step 7: Testing the Model on the Test Set
y_pred_test = rf_model.predict(X_val)
y_pred_test_labels = label_encoder.inverse_transform(y_pred_test)
y_test_labels = label_encoder.inverse_transform(y_val)
```
